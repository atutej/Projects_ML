```python
import pandas as pd
import numpy as np
```

```python
def g(z):
    g=1/(1+np.exp(-z))
    return g
```

```python
df1=pd.read_csv("/home/atswi/Downloads/mnist_train.csv",header=None,nrows=15000)
y=df1[0]
y=y.to_numpy()
df1=df1.drop(columns=[0])
df1.insert(0,"x_0",1)
trainx=df1.to_numpy()
trainx=trainx/255
```

```python
trainy=np.zeros((len(y),10))   #One-Hot encoding
trainy[np.arange(len(y)),y]=1
trainy=trainy.transpose()
```

```python
m=15000
```

```python
trainx=trainx.T
```

```python
df2=pd.read_csv("/home/atswi/Downloads/mnist_test.csv",header=None)
yt=df2[0]
yt=yt.to_numpy()
df2=df2.drop(columns=[0])
df2.insert(0,"x_0",1)
```

```python
testx=df2.to_numpy()
testx=testx/255
testx=testx.T

testy=np.zeros((len(yt),10))   #One-Hot encoding
testy[np.arange(len(yt)),yt]=1
testy=testy.transpose()
```

```python
n_x=785
n_h=64
alpha=2

theta1=np.random.randn(n_h, n_x)
theta2=np.random.randn(10, n_h)
```

```python
for i in range(2000):

    z1=theta1.dot(trainx) 
    a1=g(z1)
    z2=theta2.dot(a1) 
    a2=g(z2)

    dz2=a2-trainy
    dtheta2=(1./m)*dz2.dot(a1.T)

    da1=theta2.T.dot(dz2)
    dz1=da1 * g(z1)*(1-g(z1))
    dtheta1=(1./m)*dz1.dot(trainx.T)

    theta2=theta2-alpha*dtheta2
    theta1 = theta1-alpha*dtheta1


print("Weights 1: ")
print(theta1)
print("Weights 2: ")
print(theta2)
```

```python
z1=theta1.dot(testx) 
a1=g(z1)
z2=theta2.dot(a1) 
a2=g(z2)

predict=np.argmax(a2,axis=0)
print(predict)
```

```python
error=0
for i in range(len(predict)):
  if(predict[i]!=yt[i]):
    error=error+1
acc=(len(predict)-error)/len(predict)*100
print("Accuracy: "+str(acc)+"%")
```
